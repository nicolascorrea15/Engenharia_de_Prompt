```python
'''
Crie um script Python simples para verificar a temperatura de um sistema. Se for maior que 30, exiba um alerta de superaquecimento.
'''

try:
    # Solicita a temperatura ao usuário e armazena como string
    temperatura_str = input("Digite a temperatura do sistema em Celsius: ")
    # Converte a string de temperatura para um número de ponto flutuante (float)
    temperatura = float(temperatura_str)

    # Verifica se a temperatura é maior que 30 graus Celsius
    if temperatura > 30:
        # Se for maior, imprime um alerta de superaquecimento usando concatenação
        print("ALERTA: Superaquecimento detectado! Temperatura acima de 30°C. Temperatura atual: " + str(temperatura) + "°C.")
    else:
        # Caso contrário, informa que a temperatura está normal usando concatenação
        print("Temperatura normal. A temperatura atual é de " + str(temperatura) + "°C.")

except ValueError:
    # Captura o erro se a entrada do usuário não puder ser convertida em número
    print("Entrada inválida. Por favor, digite um número para a temperatura.")
```
