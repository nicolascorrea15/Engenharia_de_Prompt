```python
'''
Crie um script Python simples para verificar a temperatura de um sistema. Se for maior que 30, exiba um alerta de superaquecimento.
'''

try:
    # Solicita a temperatura ao usuário e armazena como string
    temperatura_str = input("Digite a temperatura do sistema em Celsius: ")
    # Converte a string de temperatura para um número de ponto flutuante (float)
    temperatura = float(temperatura_str)

    # Verifica se a temperatura é maior que 30 graus Celsius
    if temperatura > 30:
        # Se for maior, imprime um alerta de superaquecimento usando concatenação
        print("ALERTA: Superaquecimento detectado! Temperatura acima de 30°C. Temperatura atual: " + str(temperatura) + "°C.")
    else:
        # Caso contrário, informa que a temperatura está normal usando concatenação
        print("Temperatura normal. A temperatura atual é de " + str(temperatura) + "°C.")

except ValueError:
    # Captura o erro se a entrada do usuário não puder ser convertida em número
    print("Entrada inválida. Por favor, digite um número para a temperatura.")
```

Criar um programa que organize arquivos
baseados em pastas automaticamente,
baseando-se apenas na extensão do arquivo(ex: PDF, JPG,IMG,TXT)
Use biblioteca OS ou shutil.

```python
import os # Importa o módulo os para interagir com o sistema operacional (caminhos, diretórios)
import shutil # Importa o módulo shutil para operações de alto nível em arquivos (mover, copiar)

def organize_files_by_extension(source_directory):
    """
    Organiza arquivos em um diretório especificado, movendo-os para subpastas
    criadas com base em suas extensões. Por exemplo, 'documento.pdf' irá para
    a pasta 'PDF'. Arquivos sem extensão irão para 'SEM_EXTENSAO'.

    Args:
        source_directory (str): O caminho do diretório a ser organizado.
    """

    # Verifica se o diretório de origem existe. Se não existir, imprime um erro e sai.
    if not os.path.isdir(source_directory):
        print(f"Erro: O diretório '{source_directory}' não existe.")
        return

    print(f"Iniciando a organização dos arquivos em: {source_directory}")

    # Itera sobre todos os itens (arquivos e subdiretórios) no diretório de origem
    for filename in os.listdir(source_directory):
        # Constrói o caminho completo para o item atual
        file_path = os.path.join(source_directory, filename)

        # Ignora se o item for um diretório para evitar processar subpastas como arquivos
        if os.path.isdir(file_path):
            continue

        # Separa o nome do arquivo da sua extensão. Ex: 'documento.pdf' -> ('documento', '.pdf')
        name, extension = os.path.splitext(filename)

        # Determina o nome da pasta de destino com base na extensão
        if extension: # Se o arquivo tem uma extensão (ex: .pdf, .jpg)
            # Remove o ponto inicial da extensão e converte para maiúsculas (ex: '.pdf' -> 'PDF')
            extension_folder_name = extension[1:].upper()
        else: # Se o arquivo não tem extensão
            extension_folder_name = "SEM_EXTENSAO" # Define uma pasta padrão para arquivos sem extensão

        # Constrói o caminho completo para a pasta de destino
        destination_folder = os.path.join(source_directory, extension_folder_name)

        # Cria a pasta de destino se ela ainda não existir. 'exist_ok=True' evita erro se já existir.
        os.makedirs(destination_folder, exist_ok=True)

        # Tenta mover o arquivo para a sua respectiva pasta de extensão
        try:
            shutil.move(file_path, destination_folder)
            print(f"Movido '{filename}' para '{extension_folder_name}/'")
        except Exception as e:
            # Captura e imprime qualquer erro que ocorra durante o movimento do arquivo
            print(f"Erro ao mover '{filename}': {e}")

    print("Organização de arquivos concluída!")
```
